```python
data = [100,101,102]
series = pd.Series(data,index= ["a","b","c"])
series
```

```python
print(series.loc["a"])
```

```python
print(series.loc["c"])
```

```python
series.loc["c"] = 200
print(series)
```

```python
# data by index
print(series.iloc[0])
```

```python
print(series.iloc[2])
```

```python
data = [100,101,102,200,202]
series = pd.Series(data,index= ["a","b","c","d","e"])
series
```

```python
print(series[series >= 200])
```

```python
print(series[series < 200])
```
