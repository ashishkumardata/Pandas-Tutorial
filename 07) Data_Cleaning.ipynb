```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.head(5)
```

```python
# droping irrelevant column
df = df.drop(columns=["Legendary","No"])
df.head(5)
```

```python
# handling missing value
df = df.dropna(subset=["Type2"])
print(df.to_string())
print(df["Name"].count())
```

```python
df = pd.read_csv("data.csv")
df = df.fillna({"Type2":"Non"})
print(df.to_string())
```

```python
# fix inconsistent values
df["Type1"] = df["Type1"].replace({"Grass":"GRASS"})
print(df.to_string())
```

```python
df["Type1"] = df["Type1"].replace({"Fire":"GRASS",
                                   "Water":"WATER"})
print(df.to_string())
```

```python
# STANDARDIZE TEXT
df["Name"] = df["Name"].str.lower()
print(df)
```

```python
# FIX DATATYPES
df["Legendary"] = df["Legendary"].astype(bool)
df
```

```python
# REMOVE DUPLICATES VALUES
df = df.drop_duplicates()
df
```
