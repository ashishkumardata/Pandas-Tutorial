```python
import pandas as pd
```

```python
data = {"Name": ["Spongebob","Patrick","Squidward"],
        "Age":[30,35,50]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index= ["Employee1","Employee2","Employee3"])
df
```

```python
df.loc["Employee1"]
```

```python
df.iloc[2]
```

```python
# Add a new column
df["Job"] = ["Cook","N/A","Cashier"]
df
```

```python
# Add a new row
new_row = pd.DataFrame([{"Name":"Sandy","Age":28,"Job":"Engineer"}],
                       index=["Employee4"])
df = pd.concat([df, new_row])
df
```

```python
# Add new rows
new_rows = pd.DataFrame([{"Name":"Eugene","Age":60,"Job":"Manager"},
                        {"Name":"Harry","Age":29,"Job":"Assistant"}],
                       index=["Employee5","Employee6"])
df = pd.concat([df, new_rows])
df
```
